```python
import smtplib
import random
import tkinter as tk
from tkinter import messagebox

# Global variables to track the number of attempts and generated OTP
attempts = 0
generated_otp = ''

def generate_otp():
    return ''.join(random.choices('0123456789', k=6))

def send_otp(email, otp):
    global attempts
    smtp_server = 'smtp.gmail.com'
    smtp_port = 587 
    sender_email = 'abc@gmail.com'
    sender_password = '-----'

    subject = 'Your One Time Password (OTP)'
    body = f'Your OTP is: {otp}'

    message = f'Subject: {subject}\n\n{body}'

    try:
        server = smtplib.SMTP(smtp_server, smtp_port)
        server.starttls()
        server.login(sender_email, sender_password)
        server.sendmail(sender_email, email, message)
        messagebox.showinfo("OTP Sent", "OTP has been sent to your email address.")
    except Exception as e:
        messagebox.showerror("Error", f"Error: {e}")
    finally:
        server.quit()

def verify_otp(otp_input, otp_generated):
    return otp_input == otp_generated

def verify_and_show_result():
    global attempts
    global generated_otp
    entered_otp = otp_entry.get()
    if verify_otp(entered_otp, generated_otp):
        messagebox.showinfo("OTP Verified", "OTP verification successful!")
    else:
        messagebox.showerror("OTP Verification Failed", "OTP verification failed. Please try again.")
        attempts += 1
        if attempts >= 3:
            messagebox.showerror("Too Many Attempts", "You have exceeded the maximum number of attempts. Please enter your email details again.")
            attempts = 0  # Reset attempts counter
            email_entry.delete(0, 'end')
            otp_entry.delete(0, 'end')
            generated_otp = ''

def send_otp_and_show_result():
    global generated_otp
    global attempts
    if attempts < 3:
        generated_otp = generate_otp()
        send_otp(email_entry.get(), generated_otp)
    else:
        messagebox.showerror("Too Many Attempts", "You have exceeded the maximum number of attempts. Please enter your email details again.")
        attempts = 0  # Reset attempts counter
        email_entry.delete(0, 'end')
        otp_entry.delete(0, 'end')
        generated_otp = ''

# Create GUI window
root = tk.Tk()
root.title("OTP Verification")

# Email Label and Entry
email_label = tk.Label(root, text="Enter your email address:")
email_label.pack()
email_entry = tk.Entry(root)
email_entry.pack()

# Button to send OTP
send_otp_button = tk.Button(root, text="Send OTP", command=send_otp_and_show_result)
send_otp_button.pack()

# OTP Label and Entry
otp_label = tk.Label(root, text="Enter the OTP you received:")
otp_label.pack()
otp_entry = tk.Entry(root)
otp_entry.pack()

# Button to verify OTP
verify_otp_button = tk.Button(root, text="Verify OTP", command=verify_and_show_result)
verify_otp_button.pack()

root.mainloop()
```
